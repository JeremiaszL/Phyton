```python
number = 7

if number % 2 == 0:
    print(f"Liczba {number} jest parzysta")
else:
    print(f"Liczba {number} jest nieparzysta")

print("Liczby od 1 do 10:")
for i in range(1, 11):
    print(i)

print("Pętla while (5 iteracji):")
count = 0
while count < 5:
    print(f"Iteracja {count + 1}")
    count += 1
```
