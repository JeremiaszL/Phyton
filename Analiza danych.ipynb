```python
import pandas as pd

data = { 
    "Imię": ["Ala", "Bartek", "Celina"], 
    "Wiek": [22, 25, 20], 
    "Ocena": [4.5, 3.8, 5.0] 
} 

df = pd.DataFrame(data)
display(df)

print(df.describe())

df["Rok studiów"] = [2, 3, 1]

srednia_ocena = df["Ocena"].mean()
print("\nŚrednia ocena:", round(srednia_ocena, 2))

df_wysoka_ocena = df[df["Ocena"] > 4]
print("\nWiersze, w których ocena > 4:")
display(df_wysoka_ocena)
```
